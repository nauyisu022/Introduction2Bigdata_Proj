```python
paddle.enable_static()
jieba.enable_paddle()
jieba.load_userdict('jieba/user_dict.txt')
csv_list = glob.glob('data/danmaku/*/*.csv')
len(csv_list)
csv_list[0]
```

```python
def make_word_lists(csv_file_path):
    with open(csv_file_path, 'r') as f:
        reader = csv.reader((line.replace('\0', '') for line in f))

        words = []

        for row in reader:
            try:
                x = jieba.lcut(row[0], cut_all=False)
            except Exception:
                x = []
            words += x

    return words
```

```python
words = []
for csv_file_path in csv_list[:1000]:
    words += make_word_lists(csv_file_path)
```

```python
jieba.add_word('点赞', freq=None, tag=None)
```

```python
counts = {}

repeat_cha=['哈','6','啊']

for word in words:
    if len(word) == 1:  # 单个词语不计算在内
        continue
    else:
        for cha in repeat_cha:
            if word[0] == cha:
                flag=True
                for c in word:
                    if c != cha:
                        flag=False
                        break
                if flag:
                    word=cha*3
        counts[word] = counts.get(word, 0) + 1  # 遍历所有词语，每出现一次其对应的值加 1

items = list(counts.items())  #将键值对转换成列表
items.sort(key=lambda x: x[1], reverse=True)  # 根据词语出现的次数进行从大到小排序
```
